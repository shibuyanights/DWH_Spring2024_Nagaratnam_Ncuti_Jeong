```python
import pandas as pd
import requests
from datetime import datetime

# Load the list of locations from the CSV file using a different encoding
try:
    cities_df = pd.read_csv('/Users/jenu/Downloads/ch.csv', encoding='ISO-8859-1')
except UnicodeDecodeError:
    print("Error reading the file with 'ISO-8859-1' encoding.")
    # You can try other encodings if necessary

# Extract latitudes, longitudes, and city names
locations = cities_df[['lat', 'lng', 'city']].to_dict(orient='records')

def get_weather(api_key, lat, lng):
    url = f"http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lng}&appid={api_key}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to get weather data for lat: {lat}, lng: {lng}. HTTP Status code: {response.status_code}")
        print(f"Response content: {response.content}")
        return None

# Your API key here
api_key = '1d0e2051277f90a3b30f4b7f2fb812e0'

weather_data = []

for location in locations:
    lat = location['lat']
    lng = location['lng']
    city = location['city']
    data = get_weather(api_key, lat, lng)
    if data:
        for entry in data['list']:
            weather_data.append({
                'city': city,
                'lat': lat,
                'lng': lng,
                'datetime': datetime.fromtimestamp(entry['dt']),
                'temperature_celsius': entry['main']['temp'] - 273.15,  # Convert from Kelvin to Celsius
                'weather': entry['weather'][0]['description']
            })

# Convert the weather data to a pandas DataFrame
weather_df = pd.DataFrame(weather_data)

# Save the DataFrame to a CSV file
weather_df.to_csv('5_day_3hour_weather_forecast.csv', index=False)

print("Weather data saved to 5_day_3hour_weather_forecast.csv")
```
